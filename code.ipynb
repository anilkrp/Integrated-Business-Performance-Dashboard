```python
import sqlite3
import pandas as pd

# Mapping of CSV filenames to desired SQLite table names
csv_to_table = {
    'customers.csv': 'customers',
    'departments.csv': 'departments',
    'employees.csv': 'employees',
    'orders.csv': 'orders',
    'products.csv': 'products',
    'returns.csv': 'returns'
}

# load CSV files into pandas DataFrames and check all info, statictic, and unique values, missing values, and data types, unique values, convert into SQLite tables as named of csv files and in orderflow as sql database named as "orderflow.db"
dataframes = {}
for csv_file, table_name in csv_to_table.items():
    df = pd.read_csv(csv_file)
    dataframes[table_name] = df
    print("=" * 100)
    print(f"DataFrame for {table_name}:")
    print("=" * 100)
    print("Basic Info:")
    print(df.info())
    print("-" * 100)
    print("Descriptive Statistics:")
    print(df.describe())
    print("-" * 100)
    print("Unique Values:")
    print(df.nunique())
    print("-" * 100)
    print("Missing Values:")
    print(df.isnull().sum())
    print("\n")



# Create SQLite database and insert DataFrames as tables
conn = sqlite3.connect('orderflow.db')
for table_name, df in dataframes.items():
    df.to_sql(table_name, conn, if_exists='replace', index=False)
conn.close()
print("DataFrames have been successfully inserted into the SQLite database 'orderflow.db'.")
```
